```python
import numpy as np


A = 20000 + 5000 * np.arange(11)
B = 30000 * 1.08 ** np.arange(11)
C = 10000 * 1.15 ** np.arange(11)

portfolio = np.array([])

for i in range(len(A)):
    portfolio = np.append(portfolio, round(A[i] + B[i] + C[i], 2))

year_up_c = None

for i in range(100):
    new_c = 10000 * 1.15 ** np.arange(i+1)
    new_a = 20000 + 5000 * np.arange(i+1)
    if new_c[i] > new_a[i]:
        year_up_c = i
        break
    
print(A[10], B[10], C[10])
print(portfolio)
print(round(portfolio[10] - portfolio[0],2), "Разница между суммой портфеля за 10 лет")
print(year_up_c, "лет актив C впервые превысит актив A")
print(round(np.linalg.norm(portfolio, ord=1), 2), "L1 - сумма всех изменений портфеля за 10 лет")
print(round(np.linalg.norm([A[10], B[10], C[10]]), 2), "L2 - хз")
```

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(11)
A = 20000 + 5000 * years
B = 30000 * (1.08 ** years)
C = 10000 * (1.15 ** years)
portfolio = A + B + C

plt.figure(figsize=(10, 6))
plt.plot(years, A, label='Актив A (линейный)', marker='o')
plt.plot(years, B, label='Актив B (8%)', marker='s')
plt.plot(years, C, label='Актив C (15%)', marker='^')
plt.plot(years, portfolio, label='Портфель', linewidth=3, color='black')

plt.title('Рост инвестиционного портфеля')
plt.xlabel('Год')
plt.ylabel('Сумма ($)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math as m

v1 = np.array([1, 2, 3])
v2 = np.array([7, 8, 9])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

v3 = np.array([m.sqrt((v2[0] - v1[0])**2), m.sqrt((v2[1] - v1[1])**2), m.sqrt((v2[2] - v1[2])**2)])

# Рисуем векторы из начала координат
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label='v1')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label='v2')
ax.quiver(v1[0], v1[1], v1[2], v3[0], v3[1], v3[2], color='g', label='v2')

# Настройки
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_zlim([0, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()
plt.title('Векторы в 3D')
plt.show()
```

```python
import numpy as np

# Пример входных данных (у тебя в коде зашиты, но в реальной задаче читаешь с input())
# "3 3 2 1" → вектор a: начало (3, 3), конец (2, 1)
# "2 -3 8 -2" → вектор b: начало (2, -3), конец (8, -2)

data_a = np.array("3 3 2 1".split(), dtype=float)
data_b = np.array("2 -3 8 -2".split(), dtype=float)

# Извлекаем начало и конец
start_a = data_a[:2]   # (3, 3)
end_a = data_a[2:]     # (2, 1)
vec_a = end_a - start_a  # (2 - 3, 1 - 3) = (-1, -2)

start_b = data_b[:2]   # (2, -3)
end_b = data_b[2:]     # (8, -2)
vec_b = end_b - start_b  # (8 - 2, -2 - (-3)) = (6, 1)

# Сумма векторов
vec_sum = vec_a + vec_b  # (-1 + 6, -2 + 1) = (5, -1)

# Модуль (длина) вектора суммы
modulus = np.linalg.norm(vec_sum)

print(modulus)
```
